```python
import pandas as pd
df=pd.read_excel('/content/data.xlsx')
```

```python
from geopy.geocoders import Nominatim
def get_state_with_geocoding(location):
    geolocator = Nominatim(user_agent="my_geocoder")

    try:
        # Perform geocoding to get location details
        location_info = geolocator.geocode(location, addressdetails=True)

        # Extract state information
        if location_info and 'address' in location_info.raw:
            state = location_info.raw['address'].get('state', 'Not found')
            return state
        else:
            return 'Not found'

    except Exception as e:
        return f"An error occurred: {str(e)}"
```

```python
df['locations']=df['user_location'].apply(get_state_with_geocoding)
print(df)
```

```python
excel_file_path = 'data1.xlsx'
df.to_excel(excel_file_path, index=False)
```
